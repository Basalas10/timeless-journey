```python
# 3. Import Required Libraries
from pycocotools.coco import COCO
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import numpy as np
import torch
from sklearn.model_selection import train_test_split
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM
```

```python
# 4. Load a Subset of COCO Dataset
annotation_file = 'http://images.cocodataset.org/annotations/annotations_trainval2017.zip'
!wget -q $annotation_file
!unzip -o annotations_trainval2017.zip

coco_caps = COCO('annotations/captions_val2017.json')
```

```python
# 5. Data Cleaning (Using Only 1000 Samples to Reduce Memory Usage)
subset_size = 1000

captions = pd.DataFrame(coco_caps.dataset['annotations'][:subset_size])
captions = captions[['image_id', 'caption']]
captions.drop_duplicates(subset='caption', inplace=True)
captions.dropna(subset=['caption'], inplace=True)
captions = captions[captions['caption'].str.split().apply(len) >= 3]
print(f"Using a reduced dataset with {len(captions)} samples.")
```

```python
# 6. Feature Engineering
captions['caption_length'] = captions['caption'].apply(lambda x: len(x.split()))
plt.figure(figsize=(10, 5))
sns.histplot(captions['caption_length'], bins=20, kde=True)
plt.title("Caption Length Distribution")
plt.show()
```

```python
# 7. Data Splitting
train_data, test_data = train_test_split(captions, test_size=0.2, random_state=42)
val_data, test_data = train_test_split(test_data, test_size=0.5, random_state=42)
print(f"Training size: {len(train_data)}")
print(f"Validation size: {len(val_data)}")
print(f"Test size: {len(test_data)}")
```

```python
# 8. Data Quality Metrics
print("\n--- Data Quality Metrics ---")
print("Completeness:", train_data.notnull().mean())
print("Caption Length Stats:", train_data['caption_length'].describe())
```

```python
# 9. Initial Model Implementation (Using Smaller T5 Model Instead of RAG to Reduce Disk Usage)
device = "cuda" if torch.cuda.is_available() else "cpu"
print(f"\nRunning on device: {device}")

model_name = "t5-small"

tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForSeq2SeqLM.from_pretrained(model_name).to(device)

sample_text = train_data['caption'].iloc[0]
inputs = tokenizer(sample_text, return_tensors="pt").to(device)
output_ids = model.generate(**inputs)
output_text = tokenizer.decode(output_ids[0], skip_special_tokens=True)

print("\n--- Initial Output ---")
print("Sample Caption:", sample_text)
print("Generated Output:", output_text)
```
