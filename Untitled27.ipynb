```python
# ------------------------------
# 1. Setup & Data Collection
# ------------------------------
import yfinance as yf
import pandas as pd
import numpy as np
import json
import matplotlib.pyplot as plt
!pip install PyPortfolioOpt

# For FRED data
from fredapi import Fred
fred_api_key = '48c29c9bdd67e391a0385b303f5db669'
fred = Fred(api_key=fred_api_key)

# For Alpha Vantage (if needed)
# from alpha_vantage.timeseries import TimeSeries
# av_api_key = 'OTO1YOGFKXFIDMBD'
# ts_av = TimeSeries(key=av_api_key, output_format='pandas')

# Define a list of S&P500 tickers (here a subset; extend to 150 as needed)
tickers = ["AAPL", "MSFT", "NVDA", "AMZN", "META", "BRK-B", "GOOGL", "AVGO", "GOOG", "TSLA",
    "JPM", "LLY", "V", "XOM", "UNH", "MA", "NFLX", "COST", "PG", "JNJ",
    "WMT", "HD", "ABBV", "BAC", "CVX", "KO", "CRM", "CSCO", "PM", "WFC",
    "ORCL", "ABT", "IBM", "MCD", "MRK", "LIN", "GE", "PEP", "T", "TMO",
    "ACN", "VZ", "PLTR", "DIS", "ISRG", "RTX", "QCOM", "GS", "AMD", "ADBE",
    "INTU", "NOW", "TXN", "AMGN", "CAT", "PGR", "SPGI", "BKNG", "UBER", "AXP",
    "BSX", "MS", "NEE", "UNP", "PFE", "CMCSA", "BLK", "HON", "GILD", "SCHW",
    "TJX", "C", "BA", "DHR", "COP", "LOW", "VRTX", "TMUS", "SYK", "FI",
    "ADP", "DE", "AMAT", "MMC", "BMY", "PANW", "MDT", "SBUX", "CB", "ETN",
    "BX", "ADI", "PLD", "INTC", "MU", "ICE", "ELV", "AMT", "MO", "SO",
    "LRCX", "CME", "KLAC", "LMT", "WELL", "DUK", "CRWD", "CI", "MDLZ", "AJG",
    "AON", "CVS", "WM", "ECL", "MCK", "APH", "ORLY", "ANET", "UPS", "MMM",
    "PH", "EQIX", "SHW", "KKR", "TT", "NKE", "TDG", "CL", "ZTS", "MCO",
    "WMB", "MSI", "CDNS", "EOG", "CTAS", "NOC", "PNC", "CMG", "SNPS", "GD",
    "PYPL", "COF", "USB", "ITW", "REGN", "DASH", "BDX", "APD", "CEG", "AZO"]

# Download historical stock data using yfinance
# We download data from 2020-01-01 to 2025-01-01
stock_data = yf.download(tickers, start="2020-01-01", end="2025-01-01", group_by='ticker')

# For demonstration, we use AAPL data
stock_df = stock_data['AAPL'].copy()
stock_df.index = pd.to_datetime(stock_df.index)

# Get fundamental info for AAPL and print nicely formatted JSON
aapl = yf.Ticker('AAPL')
aapl_info = aapl.info
print("AAPL Fundamental Info:")
print(json.dumps(aapl_info, indent=2))

# Download FRED GDP data (quarterly) and filter from 2020 to 2025
gdp = fred.get_series('GDP')
gdp.index = pd.to_datetime(gdp.index)
start_date = "2020-01-01"
end_date = "2025-12-31"
gdp_filtered = gdp[start_date:end_date]
gdp_df = pd.DataFrame(gdp_filtered, columns=['GDP'])
print("\nFRED GDP Data:")
print(gdp_df)
```

```python
# ------------------------------
# 2. Data Preprocessing & Feature Engineering
# ------------------------------

# Upsample the quarterly GDP data to daily frequency (forward-fill values)
gdp_daily = gdp_df.resample('D').ffill()

# Merge AAPL stock data with daily GDP data on the date index
combined_df = pd.merge(stock_df, gdp_daily, left_index=True, right_index=True, how='left')
combined_df = combined_df.ffill()  # Use forward-fill

# Feature Engineering on AAPL Data:
# Calculate log returns based on the 'Close' price
combined_df['Log_Return'] = np.log(combined_df['Close'] / combined_df['Close'].shift(1))

# 20-day moving average of the Close price
combined_df['MA20'] = combined_df['Close'].rolling(window=20).mean()

# 20-day rolling volatility (standard deviation)
combined_df['Volatility'] = combined_df['Close'].rolling(window=20).std()

# Define the target as the next day's log return
combined_df['Target'] = combined_df['Log_Return'].shift(-1)

# Drop rows with missing values (e.g., first row and last row after shifting)
combined_df.dropna(inplace=True)
```

```python
# ------------------------------
# 3. Model Development: Random Forest & XGBoost
# ------------------------------

from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from xgboost import XGBRegressor
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

# Define features and target
features = ['Log_Return', 'MA20', 'Volatility']
X = combined_df[features]
y = combined_df['Target']

# Split the data (keeping time order; no shuffling)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, shuffle=False)

# Random Forest Model
rf_model = RandomForestRegressor(n_estimators=100, random_state=42)
rf_model.fit(X_train, y_train)
rf_preds = rf_model.predict(X_test)
print("\nRandom Forest Evaluation:")
print("MSE:", mean_squared_error(y_test, rf_preds))
print("MAE:", mean_absolute_error(y_test, rf_preds))
print("R²:", r2_score(y_test, rf_preds))

# XGBoost Model
xgb_model = XGBRegressor(n_estimators=100, random_state=42)
xgb_model.fit(X_train, y_train)
xgb_preds = xgb_model.predict(X_test)
print("\nXGBoost Evaluation:")
print("MSE:", mean_squared_error(y_test, xgb_preds))
print("MAE:", mean_absolute_error(y_test, xgb_preds))
print("R²:", r2_score(y_test, xgb_preds))
```

```python
# ------------------------------
# 4. LSTM Neural Network Model
# ------------------------------
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense

# Helper function to create sequences
def create_sequences(data, target, window_size):
    X_seq, y_seq = [], []
    for i in range(len(data) - window_size):
        X_seq.append(data[i:i+window_size])
        y_seq.append(target[i+window_size])
    return np.array(X_seq), np.array(y_seq)

window_size = 20
X_seq, y_seq = create_sequences(X.values, y.values, window_size)

# Split sequence data into training and testing sets
split_index = int(0.8 * len(X_seq))
X_train_seq, X_test_seq = X_seq[:split_index], X_seq[split_index:]
y_train_seq, y_test_seq = y_seq[:split_index], y_seq[split_index:]

# Build and compile LSTM model
model_lstm = Sequential()
model_lstm.add(LSTM(50, return_sequences=True, input_shape=(X_train_seq.shape[1], X_train_seq.shape[2])))
model_lstm.add(LSTM(50))
model_lstm.add(Dense(1))
model_lstm.compile(optimizer='adam', loss='mse')

# Train LSTM model
print("\nTraining LSTM Model...")
model_lstm.fit(X_train_seq, y_train_seq, epochs=10, batch_size=32, validation_split=0.1)

# LSTM Predictions
lstm_preds = model_lstm.predict(X_test_seq)
print("LSTM predictions shape:", lstm_preds.shape)
```

```python
# ------------------------------
# 5. Benchmarking: CAPM Estimation
# ------------------------------
from sklearn.linear_model import LinearRegression

# Fetch S&P 500 market index data (ticker: ^GSPC)
market_ticker = "^GSPC"
market_data = yf.Ticker(market_ticker).history(period="max")
market_data.index = pd.to_datetime(market_data.index)

# Compute market log returns
market_data['Market_Return'] = np.log(market_data['Close'] / market_data['Close'].shift(1))
market_returns = market_data[['Market_Return']].dropna()

# Ensure market_returns index is tz-naive:
market_returns.index = market_returns.index.tz_localize(None)

# Now merge with stock_df (which is tz-naive)
merged_df = pd.merge(stock_df[['Close']], market_returns, left_index=True, right_index=True, how='inner')
merged_df['Log_Return'] = np.log(merged_df['Close'] / merged_df['Close'].shift(1))
merged_df.dropna(inplace=True)

# CAPM: Regress AAPL log returns on market returns
capm_model = LinearRegression()
capm_model.fit(merged_df['Market_Return'].values.reshape(-1, 1), merged_df['Log_Return'].values)
beta = capm_model.coef_[0]
print("\nCAPM Estimated Beta for AAPL:", beta)
```

```python
# ------------------------------
# 6. Benchmarking: Modern Portfolio Theory (MPT)
# ------------------------------
# Note: For MPT, we need daily returns for multiple stocks.
# We extract adjusted close prices for our tickers and compute returns.
# Install pyportfolioopt if you haven't: !pip install PyPortfolioOpt

from pypfopt import expected_returns, risk_models, EfficientFrontier

# Create a DataFrame to hold adjusted close prices for all tickers
adj_close = pd.DataFrame()
for t in tickers:
    ticker_data = stock_data[t]
    # Use the 'Adj Close' column if available; otherwise fallback to 'Close'
    if 'Adj Close' in ticker_data.columns:
        adj_close[t] = ticker_data['Adj Close']
    else:
        adj_close[t] = ticker_data['Close']

# Ensure the index is datetime and filter for our period
adj_close.index = pd.to_datetime(adj_close.index)
adj_close = adj_close.loc['2020-01-01':'2025-01-01']

# Calculate expected returns and sample covariance matrix
mu = expected_returns.mean_historical_return(adj_close)
S = risk_models.sample_cov(adj_close)

# Optimize for maximal Sharpe ratio
ef = EfficientFrontier(mu, S)
weights = ef.max_sharpe()
cleaned_weights = ef.clean_weights()
print("\nOptimized Portfolio Weights using MPT:")
print(cleaned_weights)

# Get portfolio performance metrics
performance = ef.portfolio_performance(verbose=True)
```
