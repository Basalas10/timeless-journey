```python
import yfinance as yf
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import json
import requests
from fredapi import Fred

# Uncomment and run these lines if you haven't installed the required packages:
!pip install fredapi
!pip install PyPortfolioOpt
!pip install shap
!pip install lime

# ------------------------------
# 1. Setup & Data Collection
# ------------------------------

# Set up API keys
fred_api_key = '48c29c9bdd67e391a0385b303f5db669'
alpha_vantage_api_key = 'OTO1YOGFKXFIDMBD'

# Initialize FRED
fred = Fred(api_key=fred_api_key)

# Define a list of S&P 500 tickers (subset for demonstration)
tickers = [
     "AAPL", "MSFT", "NVDA", "AMZN", "META", "BRK-B", "GOOGL", "AVGO", "GOOG", "TSLA",
    "JPM", "LLY", "V", "XOM", "UNH", "MA", "NFLX", "COST", "PG", "JNJ",
    "WMT", "HD", "ABBV", "BAC", "CVX", "KO", "CRM", "CSCO", "PM", "WFC",
    "ORCL", "ABT", "IBM", "MCD", "MRK", "LIN", "GE", "PEP", "T", "TMO",
    "ACN", "VZ", "PLTR", "DIS", "ISRG", "RTX", "QCOM", "GS", "AMD", "ADBE",
    "INTU", "NOW", "TXN", "AMGN", "CAT", "PGR", "SPGI", "BKNG", "UBER", "AXP",
    "BSX", "MS", "NEE", "UNP", "PFE", "CMCSA", "BLK", "HON", "GILD", "SCHW",
    "TJX", "C", "BA", "DHR", "COP", "LOW", "VRTX", "TMUS", "SYK", "FI",
    "ADP", "DE", "AMAT", "MMC", "BMY", "PANW", "MDT", "SBUX", "CB", "ETN",
    "BX", "ADI", "PLD", "INTC", "MU", "ICE", "ELV", "AMT", "MO", "SO",
    "LRCX", "CME", "KLAC", "LMT", "WELL", "DUK", "CRWD", "CI", "MDLZ", "AJG",
    "AON", "CVS", "WM", "ECL", "MCK", "APH", "ORLY", "ANET", "UPS", "MMM",
    "PH", "EQIX", "SHW", "KKR", "TT", "NKE", "TDG", "CL", "ZTS", "MCO",
    "WMB", "MSI", "CDNS", "EOG", "CTAS", "NOC", "PNC", "CMG", "SNPS", "GD",
    "PYPL", "COF", "USB", "ITW", "REGN", "DASH", "BDX", "APD", "CEG", "AZO"
]

# Convert tickers list to a space-separated string (yfinance expects a string)
tickers_str = " ".join(tickers)

# Define the date range
start_date = "2020-01-01"
end_date = "2025-01-01"

# Download historical stock data from Yahoo Finance
stock_data = yf.download(tickers_str, start=start_date, end=end_date, group_by='ticker')

# For demonstration, work with AAPL data
aapl_df = stock_data['AAPL'].copy()
aapl_df.index = pd.to_datetime(aapl_df.index)

# Retrieve and print AAPL fundamental info (for company-level metrics)
aapl_ticker = yf.Ticker("AAPL")
aapl_info = aapl_ticker.info
print("AAPL Fundamental Info:")
print(json.dumps(aapl_info, indent=2))

# ------------------------------
# 2. Macro Data from FRED
# ------------------------------

# Retrieve GDP (quarterly), Interest Rates, and Inflation from FRED
gdp = fred.get_series('GDP')
gdp.index = pd.to_datetime(gdp.index)
gdp = gdp[start_date:end_date]
gdp_df = pd.DataFrame(gdp, columns=['GDP'])

interest_rate = fred.get_series('FEDFUNDS')
interest_rate.index = pd.to_datetime(interest_rate.index)
interest_rate = interest_rate[start_date:end_date]
interest_df = pd.DataFrame(interest_rate, columns=['Interest_Rate'])

inflation = fred.get_series('CPILFESL')
inflation.index = pd.to_datetime(inflation.index)
inflation = inflation[start_date:end_date]
inflation_df = pd.DataFrame(inflation, columns=['Inflation'])

# Upsample each macro series to daily frequency using forward-fill
gdp_daily = gdp_df.resample('D').ffill()
interest_daily = interest_df.resample('D').ffill()
inflation_daily = inflation_df.resample('D').ffill()

# ------------------------------
# 3. Combine Yahoo and FRED Data & Add Fundamentals
# ------------------------------

# Merge AAPL stock data with daily-upsampled macro data
aapl_combined = aapl_df.copy()
aapl_combined = aapl_combined.merge(gdp_daily, left_index=True, right_index=True, how='left')
aapl_combined = aapl_combined.merge(interest_daily, left_index=True, right_index=True, how='left')
aapl_combined = aapl_combined.merge(inflation_daily, left_index=True, right_index=True, how='left')
aapl_combined = aapl_combined.ffill()

# Feature Engineering: Calculate log returns, 20-day moving average, and volatility
aapl_combined['Log_Return'] = np.log(aapl_combined['Close'] / aapl_combined['Close'].shift(1))
aapl_combined['MA20'] = aapl_combined['Close'].rolling(window=20).mean()
aapl_combined['Volatility'] = aapl_combined['Close'].rolling(window=20).std()

# Incorporate company-level financial metrics (trailing P/E, EPS, revenue growth) from AAPL info
aapl_combined['trailingPE'] = aapl_info.get('trailingPE', np.nan)
aapl_combined['trailingEps'] = aapl_info.get('trailingEps', np.nan)
aapl_combined['revenueGrowth'] = aapl_info.get('revenueGrowth', np.nan)

# Define target as next day's log return and drop missing rows
aapl_combined['Target'] = aapl_combined['Log_Return'].shift(-1)
aapl_combined.dropna(inplace=True)

# ------------------------------
# 4. Scale Numeric Features
# ------------------------------

from sklearn.preprocessing import StandardScaler
scale_features = ['Log_Return', 'MA20', 'Volatility', 'GDP', 'Interest_Rate', 'Inflation',
                  'trailingPE', 'trailingEps', 'revenueGrowth']
scaler = StandardScaler()
aapl_combined[scale_features] = scaler.fit_transform(aapl_combined[scale_features])

# ------------------------------
# 5. Fetch Alpha Vantage Market Sentiment Data
# ------------------------------

sentiment_url = f"https://www.alphavantage.co/query?function=NEWS_SENTIMENT&tickers=AAPL&apikey={alpha_vantage_api_key}"
response = requests.get(sentiment_url)

if response.status_code == 200:
    sentiment_json = response.json()
    if 'feed' in sentiment_json:
        sentiment_df = pd.DataFrame(sentiment_json['feed'])
        # Convert 'time_published' using the expected format if it exists
        if 'time_published' in sentiment_df.columns:
            sentiment_df['time_published'] = pd.to_datetime(
                sentiment_df['time_published'], format='%Y%m%dT%H%M%S', errors='coerce'
            )
            sentiment_df.set_index('time_published', inplace=True)
        print("\nAlpha Vantage Market Sentiment Data (sample):")
        print(sentiment_df.head())

        # Option 1: If a specific numeric sentiment column exists:
        if 'overall_sentiment_score' in sentiment_df.columns:
            sentiment_df = sentiment_df[['overall_sentiment_score']]
        else:
            # Option 2: Otherwise, select only numeric columns
            sentiment_df = sentiment_df.select_dtypes(include=[np.number])

        # Resample sentiment data to daily frequency using mean aggregation
        sentiment_daily = sentiment_df.resample('D').mean()
        # Merge sentiment data with the combined dataset
        aapl_combined = aapl_combined.merge(sentiment_daily, left_index=True, right_index=True, how='left')
        aapl_combined = aapl_combined.ffill()
    else:
        print("No sentiment data found in Alpha Vantage response.")
else:
    print("Error fetching sentiment data:", response.status_code)

# ------------------------------
# 6. Final Combined Data Overview
# ------------------------------

print("\nCombined Data Sample:")
print(aapl_combined.head())

# ------------------------------
# 7. Modeling & Analysis (Examples)
# ------------------------------

from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from xgboost import XGBRegressor
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

# Define feature set and target
features = ['Log_Return', 'MA20', 'Volatility', 'GDP', 'Interest_Rate', 'Inflation',
            'trailingPE', 'trailingEps', 'revenueGrowth']
X = aapl_combined[features]
y = aapl_combined['Target']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, shuffle=False)

# Random Forest Model
rf_model = RandomForestRegressor(n_estimators=100, random_state=42)
rf_model.fit(X_train, y_train)
rf_preds = rf_model.predict(X_test)
print("\nRandom Forest Evaluation:")
print("MSE:", mean_squared_error(y_test, rf_preds))
print("MAE:", mean_absolute_error(y_test, rf_preds))
print("R²:", r2_score(y_test, rf_preds))

# XGBoost Model
xgb_model = XGBRegressor(n_estimators=100, random_state=42)
xgb_model.fit(X_train, y_train)
xgb_preds = xgb_model.predict(X_test)
print("\nXGBoost Evaluation:")
print("MSE:", mean_squared_error(y_test, xgb_preds))
print("MAE:", mean_absolute_error(y_test, xgb_preds))
print("R²:", r2_score(y_test, xgb_preds))

# ------------------------------
# 8. Interpretability: SHAP & LIME for Random Forest
# ------------------------------

import shap
explainer = shap.TreeExplainer(rf_model)
shap_values = explainer.shap_values(X_test)
# Display SHAP summary plot (this will display in your notebook)
shap.summary_plot(shap_values, X_test, plot_type="bar")

from lime.lime_tabular import LimeTabularExplainer
explainer_lime = LimeTabularExplainer(X_train.values, feature_names=features, mode='regression', verbose=True)
exp = explainer_lime.explain_instance(X_test.iloc[0].values, rf_model.predict, num_features=5)
exp.show_in_notebook(show_table=True)

# ------------------------------
# 9. LSTM Neural Network Model
# ------------------------------

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense

def create_sequences(data, target, window_size):
    X_seq, y_seq = [], []
    for i in range(len(data) - window_size):
        X_seq.append(data[i:i+window_size])
        y_seq.append(target[i+window_size])
    return np.array(X_seq), np.array(y_seq)

window_size = 20
X_seq, y_seq = create_sequences(X.values, y.values, window_size)
split_index = int(0.8 * len(X_seq))
X_train_seq, X_test_seq = X_seq[:split_index], X_seq[split_index:]
y_train_seq, y_test_seq = y_seq[:split_index], y_seq[split_index:]

model_lstm = Sequential()
model_lstm.add(LSTM(50, return_sequences=True, input_shape=(X_train_seq.shape[1], X_train_seq.shape[2])))
model_lstm.add(LSTM(50))
model_lstm.add(Dense(1))
model_lstm.compile(optimizer='adam', loss='mse')

print("\nTraining LSTM Model...")
model_lstm.fit(X_train_seq, y_train_seq, epochs=10, batch_size=32, validation_split=0.1)
lstm_preds = model_lstm.predict(X_test_seq)
print("LSTM predictions shape:", lstm_preds.shape)

# ------------------------------
# 10. Benchmarking: CAPM Estimation
# ------------------------------

from sklearn.linear_model import LinearRegression

market_ticker = "^GSPC"
market_data = yf.Ticker(market_ticker).history(period="max")
market_data.index = pd.to_datetime(market_data.index)
market_data['Market_Return'] = np.log(market_data['Close'] / market_data['Close'].shift(1))
market_returns = market_data[['Market_Return']].dropna()
market_returns.index = market_returns.index.tz_localize(None)

merged_df = pd.merge(aapl_df[['Close']], market_returns, left_index=True, right_index=True, how='inner')
merged_df['Log_Return'] = np.log(merged_df['Close'] / merged_df['Close'].shift(1))
merged_df.dropna(inplace=True)

capm_model = LinearRegression()
capm_model.fit(merged_df['Market_Return'].values.reshape(-1, 1), merged_df['Log_Return'].values)
beta = capm_model.coef_[0]
print("\nCAPM Estimated Beta for AAPL:", beta)

# ------------------------------
# 11. Benchmarking: Modern Portfolio Theory (MPT)
# ------------------------------

from pypfopt import expected_returns, risk_models, EfficientFrontier

adj_close = pd.DataFrame()
for t in tickers:
    ticker_data = stock_data[t]
    if 'Adj Close' in ticker_data.columns:
        adj_close[t] = ticker_data['Adj Close']
    else:
        adj_close[t] = ticker_data['Close']

adj_close.index = pd.to_datetime(adj_close.index)
adj_close = adj_close.loc['2020-01-01':'2025-01-01']
mu = expected_returns.mean_historical_return(adj_close)
S = risk_models.sample_cov(adj_close)

ef = EfficientFrontier(mu, S)
weights = ef.max_sharpe()
cleaned_weights = ef.clean_weights()
print("\nOptimized Portfolio Weights using MPT:")
print(cleaned_weights)
performance = ef.portfolio_performance(verbose=True)

# ------------------------------
# 12. Risk Assessment: Monte Carlo Simulation
# ------------------------------

# Calculate daily portfolio returns based on MPT weights
daily_returns = adj_close.pct_change().dropna()
weights_array = np.array([cleaned_weights.get(t, 0) for t in adj_close.columns])
portfolio_daily_returns = (daily_returns * weights_array).sum(axis=1)

# Monte Carlo simulation: simulate 252-day (1 year) portfolio return paths
num_simulations = 10000
num_days = 252
simulated_end_values = []
initial_value = 1  # starting portfolio value

for i in range(num_simulations):
    simulated_path = [initial_value]
    for d in range(num_days):
        daily_return = np.random.choice(portfolio_daily_returns)
        simulated_path.append(simulated_path[-1] * (1 + daily_return))
    simulated_end_values.append(simulated_path[-1])

simulated_end_values = np.array(simulated_end_values)
var_5 = np.percentile(simulated_end_values, 5)
print(f"\nMonte Carlo Simulation: 5% VaR over 1 year (starting value $1): {var_5:.4f}")

# ------------------------------
# 13. Visualization Example
# ------------------------------

plt.figure(figsize=(12, 6))
plt.plot(y_test.index, y_test, label="Actual Target", marker='o', linestyle='--')
plt.plot(y_test.index, rf_preds, label="RF Predictions", marker='x')
plt.plot(y_test.index, xgb_preds, label="XGB Predictions", marker='^')
plt.xlabel("Date")
plt.ylabel("Next Day Log Return (Scaled)")
plt.title("Model Predictions vs. Actual Returns")
plt.legend()
plt.show()

plt.figure(figsize=(12, 6))
plt.plot(range(len(y_test_seq)), y_test_seq, label="Actual LSTM Target", marker='o', linestyle='--')
plt.plot(range(len(lstm_preds)), lstm_preds, label="LSTM Predictions", marker='x')
plt.xlabel("Sample Index")
plt.ylabel("Next Day Log Return (Scaled)")
plt.title("LSTM Predictions vs. Actual Targets")
plt.legend()
plt.show()
```
