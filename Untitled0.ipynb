```python
# Install necessary packages (if not already installed)
!pip install pycocotools matplotlib seaborn pandas

# Import necessary libraries
from pycocotools.coco import COCO
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import requests
from PIL import Image
from io import BytesIO
import numpy as np

# Set display configurations
sns.set(style="whitegrid")

# Load COCO Dataset annotations
annotation_file = 'http://images.cocodataset.org/annotations/annotations_trainval2017.zip'

# Download and unzip annotations (if not already done)
!wget $annotation_file
!unzip -q annotations_trainval2017.zip

# Load COCO dataset annotations
coco = COCO('annotations/instances_val2017.json')

# 1. Display basic dataset information
print(f"Total Images: {len(coco.imgs)}")
print(f"Total Categories: {len(coco.cats)}")
print(f"Total Annotations: {len(coco.anns)}")

# 2. Visualize object category distribution
categories = [cat['name'] for cat in coco.loadCats(coco.getCatIds())]
category_ids = coco.getCatIds()
category_counts = {cat: len(coco.getImgIds(catIds=[category_id])) for cat, category_id in zip(categories, category_ids)}

plt.figure(figsize=(12, 6))
sns.barplot(x=list(category_counts.keys()), y=list(category_counts.values()))
plt.xticks(rotation=90)
plt.title("Distribution of Object Categories in COCO Dataset")
plt.xlabel("Category")
plt.ylabel("Number of Images")
plt.show()

# 3. Analyze caption lengths (text annotations)
coco_caps = COCO('annotations/captions_val2017.json')

captions = [ann['caption'] for ann in coco_caps.loadAnns(coco_caps.getAnnIds())]
caption_lengths = [len(cap.split()) for cap in captions]

plt.figure(figsize=(10, 5))
sns.histplot(caption_lengths, bins=30, kde=True)
plt.title("Distribution of Caption Lengths in COCO Dataset")
plt.xlabel("Number of Words per Caption")
plt.ylabel("Frequency")
plt.show()

# 4. Display sample images with captions
def display_sample_images(coco, coco_caps, num_samples=3):
    img_ids = coco.getImgIds()[:num_samples]
    for img_id in img_ids:
        img_data = coco.loadImgs(img_id)[0]
        response = requests.get(img_data['coco_url'])
        img = Image.open(BytesIO(response.content))
        anns = coco_caps.loadAnns(coco_caps.getAnnIds(imgIds=img_id))

        plt.imshow(img)
        plt.axis('off')
        plt.title("\\n".join([ann['caption'] for ann in anns]))
        plt.show()

display_sample_images(coco, coco_caps)

# 5. Analyze object co-occurrence matrix with corrected indexing
cat_id_to_index = {cat_id: idx for idx, cat_id in enumerate(category_ids)}
co_occurrence = np.zeros((len(category_ids), len(category_ids)))

for img_id in coco.getImgIds():
    ann_ids = coco.getAnnIds(imgIds=[img_id])
    anns = coco.loadAnns(ann_ids)
    cats_in_img = list(set([cat_id_to_index[ann['category_id']] for ann in anns if ann['category_id'] in cat_id_to_index]))
    for i in range(len(cats_in_img)):
        for j in range(len(cats_in_img)):
            co_occurrence[cats_in_img[i]][cats_in_img[j]] += 1

plt.figure(figsize=(12, 10))
sns.heatmap(co_occurrence, xticklabels=categories, yticklabels=categories, cmap="YlGnBu", cbar=True)
plt.title("Object Co-occurrence Matrix in COCO Dataset")
plt.xlabel("Object Category")
plt.ylabel("Object Category")
plt.show()

# Summary of key findings
print("\nExploratory Data Analysis Completed:")
print("- Visualized object category distribution.")
print("- Analyzed caption length distribution.")
print("- Displayed sample images with captions.")
print("- Generated object co-occurrence heatmap.")
```
